```python
import numpy as np
import pandas as pd

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def clipped_normal(rng, mean, sd, low, high, size):
    x = rng.normal(mean, sd, size)
    return np.clip(x, low, high)

def bernoulli_from_logit(rng, logit_p):
    return rng.binomial(1, sigmoid(logit_p))

# =========================================================
# 1) CATEGORIZATION FUNCTIONS
# =========================================================
def categorize_education(edu_code: np.ndarray) -> np.ndarray:
    mapping = {
        0: "0", # Primary School
        1: "1", # Junior High School
        2: "2", # Senior High School
        3: "3", # University
    }
    return np.vectorize(mapping.get)(edu_code)

def categorize_marital(married_bin: np.ndarray) -> np.ndarray:
    return np.where(married_bin == 1, "1", "0")  # 1 = married, 0 = single

def categorize_phq9(phq9: np.ndarray) -> np.ndarray:
    cat = np.empty(phq9.shape, dtype=object)
    cat[(phq9 <= 6)] = "0"  # Minimal (1-4)  (includes 0)
    cat[(phq9 >= 7) & (phq9 <= 12)] = "1"   # Mild (5-9)
    cat[(phq9 >= 13) & (phq9 <= 18)] = "2" # Moderate (10-14)
    cat[(phq9 >= 19)] = "3"               # Severe (20-27)
    return cat

def categorize_nutrition(score_0_30: np.ndarray) -> np.ndarray:
    cat = np.empty(score_0_30.shape, dtype=object)
    cat[score_0_30 >= 24] = "0"  # Normal nutritional status (24-30)
    cat[(score_0_30 >= 17) & (score_0_30 <= 23)] = "1"  # At risk (17-23.5)
    cat[score_0_30 <= 16] = "2"  # Malnourished (<17)
    return cat

def adl_category(adl: np.ndarray) -> np.ndarray:
    cat = np.empty(adl.shape, dtype=object)
    cat[adl <= 4] = "0"  # Totally dependence (0-4)
    cat[(adl >= 5) & (adl <= 8)] = "1"  # Severe dependence (5-8)
    cat[(adl >= 9) & (adl <= 11)] = "2" # Moderate dependence (9-11)
    cat[adl >= 12] = "3" # Independence (12-20)
    return cat

# =========================================================
# 2) HEALTH INSURANCE CATEGORIES
#   1 = Universal Coverage (UC / "Gold Card")
#   2 = Social Security Scheme (SSS)
#   3 = Civil Servant Medical Benefit Scheme (CSMBS) / Direct billing
#   4 = Other
# =========================================================
def generate_insurance_code(rng, n: int) -> np.ndarray:
    """
    Returns insurance_code as strings: "1","2","3","4"
    """
    probs = [0.55, 0.20, 0.20, 0.05]  # UC, SSS, CSMBS/Direct, Other
    codes = rng.choice([1, 2, 3, 4], size=n, p=probs)
    return codes.astype(str)

def generate_synthetic_stroke(
    n: int,
    seed: int = 42,
    nb_n: int = 5,
    beta = None,
) -> pd.DataFrame:

    rng = np.random.default_rng(seed)
    
        # --- load betas ---
    b = DEFAULT_BETA.copy()
    if beta is not None:
        b.update(beta)

    # --- unpack (so your existing equation works) ---
    beta_anchor_dis   = b["beta_anchor_dis"]
    beta_anchor_const = b["beta_anchor_const"]

    beta_age_dec      = b["beta_age_dec"]
    beta_female       = b["beta_female"]
    beta_low_income   = b["beta_low_income"]
    beta_low_edu      = b["beta_low_edu"]
    beta_married      = b["beta_married"]
    beta_caregiver    = b["beta_caregiver"]
    beta_smoking      = b["beta_smoking"]
    beta_insurance_low= b["beta_insurance_low"]

    beta_ht           = b["beta_ht"]
    beta_dm           = b["beta_dm"]
    beta_dlp          = b["beta_dlp"]
    beta_af           = b["beta_af"]

    beta_nihss         = b["beta_nihss"]
    beta_nihss_over15  = b["beta_nihss_over15"]
    beta_nihss_over20  = b["beta_nihss_over20"]
    beta_nihss_step15  = b["beta_nihss_step15"]

    beta_sbp_high     = b["beta_sbp_high"]
    beta_dbp_high     = b["beta_dbp_high"]
    beta_shr_t3       = b["beta_shr_t3"]

    beta_any_pna      = b["beta_any_pna"]
    beta_uti          = b["beta_uti"]
    beta_psore        = b["beta_psore"]
    beta_sepsis       = b["beta_sepsis"]
    beta_vte          = b["beta_vte"]
    beta_delir        = b["beta_delir"]
    beta_hf           = b["beta_hf"]
    beta_large_inf    = b["beta_large_inf"]
    beta_hemo_tx      = b["beta_hemo_tx"]
    beta_herni        = b["beta_herni"]
    beta_neuro        = b["beta_neuro"]

    beta_depression   = b["beta_depression"]
    beta_nut_risk     = b["beta_nut_risk"]
    beta_nut_mal      = b["beta_nut_mal"]

    beta_rtpa         = b["beta_rtpa"]
    beta_thrombectomy = b["beta_thrombectomy"]
    beta_early_rehab  = b["beta_early_rehab"]
    beta_early_mobil  = b["beta_early_mobil"]

    noise_sd          = b["noise_sd"]    

    # -----------------------------
    # Demographics / SES
    # -----------------------------
    sex_male = rng.binomial(1, 0.55, n)
    sex = np.where(sex_male == 1, "1", "0")  # optional: 1=male,0=female (keep if you want)
    # If you prefer Male/Female, replace above with: sex = np.where(sex_male==1,"Male","Female")

    age = np.rint(clipped_normal(rng, 72, 12, 18, 100, n)).astype(int)  # integer
    age_cat = np.where(age < 75, "0", "1")  # 0=<75, 1=>75 (optional helper)

    income_q_num = rng.integers(1, 6, n)
    socioeconomic = income_q_num.astype(str)  # "1".."5" instead of "Q1..Q5" (if you prefer Q, change here)

    edu_code = rng.choice([0, 1, 2, 3], size=n, p=[0.45, 0.30, 0.17, 0.08])
    education = categorize_education(edu_code)  # "0".."3"

    married = bernoulli_from_logit(rng, -0.6 + 0.05*(age-40))
    marital_status = categorize_marital(married)  # "1" married / "0" single

    caregiver = rng.binomial(1, 0.60, n)
    caregiver_cat = np.where(caregiver == 1, "1", "0")

    # ---- Health insurance (multiclass) ----
    insurance_code = generate_insurance_code(rng, n)  # "1","2","3","4"
    # Optional: an "access" proxy for treatment (CSMBS/Direct often higher access)
    insurance_access = np.select(
        [insurance_code == "3", insurance_code == "2", insurance_code == "1", insurance_code == "4"],
        [1.0, 0.7, 0.6, 0.5],
        default=0.6
    )

    # Smoking
    logit_smoke = -1.1 + 0.7*sex_male - 0.15*(income_q_num-3)
    smoking = bernoulli_from_logit(rng, logit_smoke)
    smoking_cat = np.where(smoking == 1, "1", "0")

    # -----------------------------
    # Comorbidities
    # -----------------------------
    ht = bernoulli_from_logit(rng, -0.2 + 0.035*(age-65) + 0.15*sex_male)
    dm = bernoulli_from_logit(rng, -0.7 + 0.03*(age-60) + 0.10*(income_q_num<=2))
    dlp = bernoulli_from_logit(rng, -0.3 + 0.02*(age-60))
    af = bernoulli_from_logit(rng, -2.2 + 0.05*(age-65))

    ht_cat = np.where(ht == 1, "1", "0")
    dm_cat = np.where(dm == 1, "1", "0")
    dlp_cat = np.where(dlp == 1, "1", "0")
    af_cat = np.where(af == 1, "1", "0")

    # -----------------------------
    # Acute physiology (SBP/DBP), HbA1c/glucose
    # -----------------------------
    sbp = clipped_normal(rng, 155 + 10*ht, 22, 80, 240, n)
    dbp = clipped_normal(rng, 88 + 6*ht, 14, 40, 140, n)

    hba1c = clipped_normal(rng, 5.8 + 1.2*dm, 0.9, 4.0, 14.0, n)
    eAG = 28.7*hba1c - 46.7
    glucose = clipped_normal(rng, 120 + 55*dm, 35, 50, 500, n)

    # -----------------------------
    # NIHSS severity
    # -----------------------------
    sev_shift = (
        0.02*(age-70)
        + 0.35*af
        + 0.08*(income_q_num==1)
        + 0.05*(income_q_num==2)
        + 0.10*ht
        + 0.10*dm
        + 0.05*dlp
        + rng.normal(0, 1.5, n)
    )
    
    base = rng.negative_binomial(n=5, p=0.52, size=n)
    nihss_base = np.clip(np.round(base + sev_shift).astype(int), 0, 42)
    
    """
    logit_lvo = (
        -2.2
        + 0.03*(age-70)
        + 0.8*af
        + 0.25*dm
        + 0.20*ht
    )
    lvo = bernoulli_from_logit(rng, logit_lvo)
    bump = rng.integers(12, 19, size=n) * lvo
    nihss = np.clip(nihss_base + bump, 0, 42)
    """
    logit_lvo = (
        -2.2
        + 0.03*(age-70)
        + 0.8*af
        + 0.25*dm
        + 0.20*ht
    )
    lvo = bernoulli_from_logit(rng, logit_lvo)

    # 1) Majority get a MODERATE bump -> fills 16–20
    # 2) Minority get a BIG bump -> ensures ≥21 stays 5–8%
    big = rng.binomial(1, 0.18, size=n)  # 25% of LVO are "big"

    bump_small = rng.integers(12, 16, size=n)     # +6..+10 (push into 16–20)
    bump_big   = rng.integers(15, 21, size=n)    # +12..+18 (push into ≥21)

    bump = lvo * np.where(big == 1, bump_big, bump_small)
    nihss = np.clip(nihss_base + bump, 0, 42)

    # stress response
    sbp = np.clip(sbp + 0.7*nihss, 80, 240)
    dbp = np.clip(dbp + 0.25*nihss, 40, 140)
    glucose = np.clip(glucose + 3.0*nihss, 50, 500)

    sbp_int = np.rint(sbp).astype(int)
    dbp_int = np.rint(dbp).astype(int)

    sbp_cat = np.where(sbp_int < 185, "0", "1")  # 0=<185,1=>185
    dbp_cat = np.where(dbp_int < 110, "0", "1")  # 0=<110,1=>110
    nihss_cat = np.where(nihss < 15, "0", "1")   # 0=<15,1=>15

    # SHR tertiles as "0/1/2" or "1/2/3"? you asked T1/T2/T3 earlier, but if you want numeric:
    shr = glucose / np.clip(eAG, 50, None)
    shr_tertile_num = pd.qcut(shr, 3, labels=[0, 1, 2]).astype(int)  # 0/1/2
    shr_cat = shr_tertile_num.astype(str)  # "0","1","2" (T1/T2/T3)

    # -----------------------------
    # Large infarct proxy (for thrombectomy)
    # -----------------------------
    large_inf = bernoulli_from_logit(rng, -2.3 + 0.10*nihss + 0.4*af)
    large_inf_cat = np.where(large_inf == 1, "1", "0")

    # -----------------------------
    # Treatment (Guideline-like)
    #   - NIHSS < 4 => NO rTPA
    #   - BP threshold influences rTPA probability
    # -----------------------------
    bp_ok_for_rtpa = ((sbp_int < 185) & (dbp_int < 110)).astype(int)

    mild_stroke = (nihss < 4).astype(int)               # hard rule
    moderate_stroke = ((nihss >= 4) & (nihss <= 20)).astype(int)
    very_severe = (nihss >= 21).astype(int)

    # rTPA probability
    logit_rtpa = (
        -2.0
        + 1.3*bp_ok_for_rtpa
        + 0.9*moderate_stroke
        - 0.7*very_severe
        + 0.6*insurance_access
    )
    p_rtpa = sigmoid(logit_rtpa)
    p_rtpa = np.where(mild_stroke == 1, 0.0, p_rtpa)  # NIHSS < 4 => 0

    # thrombectomy probability (rare, higher NIHSS + large infarct proxy)
    logit_thromb = -4.0 + 0.08*nihss + 1.2*large_inf + 0.5*insurance_access
    p_thromb = sigmoid(logit_thromb)

    # mutually exclusive assignment: thrombectomy -> rTPA -> conservative
    u = rng.random(n)
    treatment_code = np.zeros(n, dtype=int)  # 0=Conservative,1=rTPA,2=Thrombectomy
    thromb_mask = (u < p_thromb)
    rtpa_mask = (~thromb_mask) & (u < (p_thromb + p_rtpa))
    treatment_code[thromb_mask] = 2
    treatment_code[rtpa_mask] = 1

    # Encode treatment as strings if you want (or keep as numeric)
    # Here: "0" Conservative, "1" rTPA, "2" Thrombectomy
    treatment = treatment_code.astype(str)

    # -----------------------------
    # Rehab / mobilization
    # -----------------------------
    early_rehab = bernoulli_from_logit(rng, 0.9 - 0.10*nihss + 0.08*caregiver)
    early_mobil = bernoulli_from_logit(rng, 0.6 - 0.08*nihss + 0.05*caregiver)
    early_rehab_cat = np.where(early_rehab == 1, "1", "0")
    early_mobil_cat = np.where(early_mobil == 1, "1", "0")

    # -----------------------------
    # Complications
    # -----------------------------
    asp = bernoulli_from_logit(rng, -2.0 + 0.10*nihss + 0.02*(age-70) - 0.30*early_mobil)
    hap = bernoulli_from_logit(rng, -2.4 + 0.08*nihss + 0.02*(age-70) - 0.25*early_mobil)
    uti = bernoulli_from_logit(rng, -2.2 + 0.06*nihss + 0.01*(age-70) - 0.15*early_mobil)
    psore = bernoulli_from_logit(rng, -3.0 + 0.10*nihss + 0.02*(age-70) - 0.35*early_mobil)
    vte = bernoulli_from_logit(rng, -3.4 + 0.07*nihss + 0.01*(age-70) - 0.25*early_mobil)
    delir = bernoulli_from_logit(rng, -2.6 + 0.09*nihss + 0.03*(age-75) - 0.15*early_rehab)
    sepsis = bernoulli_from_logit(rng, -3.6 + 0.06*nihss + 0.6*hap + 0.7*asp)
    hf = bernoulli_from_logit(rng, -3.0 + 0.03*(age-70) + 0.5*af + 0.25*ht)

    hemo_tx = bernoulli_from_logit(rng, -3.0 + 0.05*nihss + 0.55*(treatment_code==1) + 0.45*(treatment_code==2))
    herni = bernoulli_from_logit(rng, -4.2 + 0.10*nihss + 0.6*large_inf)
    neuro = bernoulli_from_logit(rng, -4.0 + 0.07*nihss + 1.0*herni + 0.5*hemo_tx)

    def b01(x): return np.where(x == 1, "1", "0")
    any_pna = ((asp == 1) | (hap == 1)).astype(int)

    # -----------------------------
    # Depression (PHQ-9) -> category codes "0..4"
    # -----------------------------
    dep_dx = bernoulli_from_logit(rng, -0.8 + 0.03*nihss + 0.25*(income_q_num==1) + 0.12*(income_q_num==2))
    phq9 = np.clip(rng.poisson(lam=3.5 + 6*dep_dx, size=n), 0, 27).astype(int)
    depression_cat = categorize_phq9(phq9)  # "0".."4"
    clinical_depression_flag = (phq9 >= 7).astype(int)  # used in outcome model

    # -----------------------------
    # Nutrition score -> category codes "0..2"
    # -----------------------------
    nut_score = 24 - 0.25*nihss - 1.0*sepsis - 0.8*any_pna + 0.3*early_rehab + rng.normal(0, 2.0, n)
    nut_score = np.clip(nut_score, 0, 30)
    nut_score = np.rint(nut_score).astype(int)
    nutrition_cat = categorize_nutrition(nut_score)  # "0","1","2"
    nut_risk_ind = ((nut_score >= 17) & (nut_score <= 23)).astype(int)
    nut_mal_ind = (nut_score <= 16).astype(int)

    # -----------------------------
    # Outcome: ADL score at 6 months via your coefficient knobs (LINEAR SCORE MODEL)
    # -----------------------------
    # 1) ADL discharge score integer 0-20 (keep your original discharge generator)
    nihss_over15 = np.maximum(nihss - 15, 0)          # 0 if <=15, then grows
    nihss_over20 = np.maximum(nihss - 20, 0)          # extra penalty for very severe
    adl_dis_raw = (
        20
        - 0.6*nihss
        - 0.22 * nihss_over15
        - 2.0*any_pna
        - 1.6*delir
        - 1.0*sepsis
        - 0.6*vte
        - 1.0*clinical_depression_flag
        - 1.2*nut_mal_ind
        - 0.6*nut_risk_ind
        + 1.8*early_rehab
        + rng.normal(0, 2.8, n)
    )
    adl_dis_raw = np.clip(adl_dis_raw, 0, 20)
    adl_dis_raw = np.rint(adl_dis_raw).astype(int)
    
    adl_dis = adl_dis_raw.copy()
    
    # --- Rebalance ADL_Discharge categories while keeping clinical ranges ---
    target = np.array([0.13, 0.24, 0.28, 0.49], dtype=float)
    target = target / target.sum()  # normalize in case it doesn't sum to 1

    # Use a severity index aligned with your discharge model
    # (higher = worse)
    
    severity = (
        0.6*nihss
        + 0.22 * nihss_over15
        + 2.0*any_pna
        + 1.6*delir
        + 1.0*sepsis
        + 0.8*psore
        + 0.6*vte
        + 1.0*clinical_depression_flag
        + 1.2*nut_mal_ind
        + 0.6*nut_risk_ind
        - 1.8*early_rehab
        + rng.normal(0, 0.5, n)  # tiny noise to break ties
    )

    # Sort by severity (worst first)
    order = np.argsort(-severity)

    n0 = int(round(target[0] * n))
    n1 = int(round(target[1] * n))
    n2 = int(round(target[2] * n))
    n3 = n - (n0 + n1 + n2)  # remainder

    idx0 = order[:n0]
    idx1 = order[n0:n0+n1]
    idx2 = order[n0+n1:n0+n1+n2]
    idx3 = order[n0+n1+n2:]

    # Assign ADL integers within each category's clinical range
    # (Uniform within range; you can swap to triangular if you want more mass near boundaries)
    adl_dis[idx0] = rng.integers(0, 5, size=len(idx0))     # 0-4
    adl_dis[idx1] = rng.integers(5, 9, size=len(idx1))     # 5-8
    adl_dis[idx2] = rng.integers(9, 12, size=len(idx2))    # 9-11
    adl_dis[idx3] = rng.integers(12, 21, size=len(idx3))   # 12-20

    # Safety
    adl_dis = np.clip(adl_dis, 0, 20).astype(int)


    adl_dis_cat = adl_category(adl_dis)  # "0".."3"
    adl_dependent = (adl_dis <= 11).astype(int)

    # 2) ADL at 6 months: use YOUR betas as effects on the SCORE (0–20)
    #    - beta0 now acts as the *score intercept* at 6 months (not a logit intercept)
    #    - Choose signs: negative betas reduce ADL (worse), positive betas improve ADL (better)
    """
    adl_6m_raw = (
        beta0
        + beta_age_dec * age_dec
        + beta_nihss * nihss
        + beta_dm * dm
        + beta_af * af
        + beta_income_q1 * income_q1
        + beta_shr_t3 * shr_t3
        + beta_pna * any_pna
        + beta_delir * delir
        + beta_early_rehab * early_rehab
        + beta_depression * clinical_depression_flag
        + beta_nutrition_risk * nut_risk_ind
        + beta_nutrition_mal * nut_mal_ind
        + beta_ht * ht
        + beta_dlp * dlp
        + beta_herni * herni
        + beta_neuro * neuro
        + rng.normal(0, 2.0, n)   # noise for 6m score
    )
    """
# --- SDOH / demographics ---
    female = (sex_male == 0).astype(int)
    low_income = ((income_q_num == 1) | (income_q_num == 2)).astype(int)

    low_edu = (edu_code <= 1).astype(int)          # <= junior high (you can adjust)
    unmarried = (married == 0).astype(int)
    no_caregiver = (caregiver == 0).astype(int)

    # Insurance: treat non-CSMBS/direct as lower access proxy (tune if you want)
    insurance_low = (insurance_code != "3").astype(int)  # 1=UC/SSS/Other vs 0=CSMBS-direct

    # --- Clinical severity ---
    age_dec = age / 10.0
    nihss_over15 = np.maximum(nihss - 15, 0)          # 0 if <=15, then grows
    nihss_over20 = np.maximum(nihss - 20, 0)          # extra penalty for very severe
    nihss_step15 = (nihss > 15).astype(int)
    sbp_high = (sbp_int >= 185).astype(int)
    dbp_high = (dbp_int >= 110).astype(int)
    shr_t3 = (shr_tertile_num == 2).astype(int)

    # --- Treatments ---
    rtpa = (treatment_code == 1).astype(int)
    thrombectomy = (treatment_code == 2).astype(int)

    # --- Complications bundles ---
    any_pna = ((asp == 1) | (hap == 1)).astype(int)   # you already have this as any_pna

    
    # If you want 6m to be anchored to discharge (recommended), uncomment this version instead:
    adl_6m_raw = (

        # Anchor
        beta_anchor_dis * adl_dis
        + beta_anchor_const

        # Demographics / SDOH
        + beta_age_dec * age_dec
        + beta_female * female
        + beta_low_income * low_income
        + beta_low_edu * low_edu
        + beta_married * (1 - unmarried)
        + beta_caregiver * (1 - no_caregiver)
        + beta_smoking * smoking
        + beta_insurance_low * insurance_low

        # Comorbidities
        + beta_ht * ht
        + beta_dm * dm
        + beta_dlp * dlp
        + beta_af * af

        # Severity / physiology
        + beta_nihss * nihss
        + beta_nihss_over15 * nihss_over15
        + beta_nihss_step15 * nihss_step15
        + beta_sbp_high * sbp_high
        + beta_dbp_high * dbp_high
        + beta_shr_t3 * shr_t3


        # Complications
        + beta_any_pna * any_pna
        + beta_uti * uti
        + beta_sepsis * sepsis
        + beta_vte * vte
        + beta_delir * delir
        + beta_hf * hf
        + beta_large_inf * large_inf
        + beta_hemo_tx * hemo_tx
        + beta_herni * herni
        + beta_neuro * neuro

        # Mood / nutrition
        + beta_depression * clinical_depression_flag
        + beta_nut_risk * nut_risk_ind
        + beta_nut_mal * nut_mal_ind

        # Treatment / rehab
        + beta_rtpa * rtpa
        + beta_thrombectomy * thrombectomy
        + beta_early_rehab * early_rehab
        + beta_early_mobil * early_mobil

        # Noise
        + rng.normal(0, noise_sd, n)
    )

    adl_6m = np.clip(adl_6m_raw, 0, 20)
    adl_6m = np.rint(adl_6m).astype(int)


    adl_6m_cat = adl_category(adl_6m)
    adl_6m_dependent = (adl_6m <= 11).astype(int)


    # nutrition 6m
    nut6 = nut_score + 2 + 0.8*early_rehab - 0.7*clinical_depression_flag - 0.3*sepsis + rng.normal(0, 2.0, n)
    nut6 = np.clip(nut6, 0, 30)
    nut6 = np.rint(nut6).astype(int)
    nutrition6_cat = categorize_nutrition(nut6)

    # -----------------------------
    # DataFrame (mostly coded strings)
    # -----------------------------
    df = pd.DataFrame({
        "Sex": sex,  # "1" male / "0" female (change if you want)
        "Age": age,
        "Age_cat": age_cat,  # optional
        "Socioeconomic": socioeconomic,  # "1".."5"
        "Education": education,          # "0".."3"
        "Martial status": marital_status,# "1" married / "0" single
        "Caregiver": caregiver_cat,      # "1" yes / "0" no
        "Smoking": smoking_cat,          # "1" yes / "0" no

        # Insurance scheme (as requested)
        "HealthInsurance": insurance_code,  # "1" UC / "2" SSS / "3" CSMBS-Direct / "4" Other

        "HT": ht_cat,
        "DM": dm_cat,
        "DLP": dlp_cat,
        "AF": af_cat,

        "SBP": sbp_int,
        "SBP_cat": sbp_cat,
        "DBP": dbp_int,
        "DBP_cat": dbp_cat,

        "NIHSS": nihss,
        "NIHSS_cat": nihss_cat,

        "SHR": shr_cat,  # "0"=T1, "1"=T2, "2"=T3

        "AspirationPneumonia": b01(asp),
        "HAP": b01(hap),
        "UTI": b01(uti),
        #"PressureSore": b01(psore),
        "Sepsis": b01(sepsis),
        "VTE": b01(vte),
        "Delirium": b01(delir),
        "HF": b01(hf),
        "LargeInfarction": b01(large_inf),
        "HemorrhagicTransformation": b01(hemo_tx),
        "BrainHerniation": b01(herni),
        #"RequireNeurosurgery": b01(neuro),

        "ClinicalDepression_PHQ9": phq9,
        "ClinicalDepression_cat": depression_cat,  # "0".."4"

        "NutritionalStatus_score": nut_score,
        "NutritionalStatus": nutrition_cat,        # "0".."2"

        "Treatment": treatment,  # "0" Conservative, "1" rTPA, "2" Thrombectomy
        "EarlyRehabilitation": early_rehab_cat,     # "1"/"0"
        "EarlyMobilization": early_mobil_cat,       # "1"/"0"

        "ADL_Discharge": adl_dis,
        "ADL_Discharge_cat": adl_dis_cat,           # "0".."3"
        "ADL_dependent_cut12": adl_dependent,       # 1/0 (int)

        "ADL_6month": adl_6m,
        "ADL_6month_cat": adl_6m_cat,               # "0".."3"
        "ADL_6month_dependent_cut12": adl_6m_dependent,   # 1/0 (int)

        "NutritionalStatus_6month_score": nut6,
        "NutritionalStatus_6month": nutrition6_cat, # "0".."2"
    })

    return df
```

```python
import numpy as np
import pandas as pd

IN_PATH  = "data_gen_almost5.csv"
OUT_PATH = "data_gen_with_ADL_items_logic_5.csv"

df = pd.read_csv(IN_PATH)

# -------------------------
# Barthel item schema (10 items, total 20)
# -------------------------
ITEMS = ["Feed", "Groom", "Transfer", "Toilet", "Mobility", "Dress", "Stairs", "Bath", "Bowel", "Bladder"]
MAXS  = np.array([2, 1, 3, 2, 3, 2, 2, 1, 2, 2], dtype=int)
MINS  = np.zeros_like(MAXS)
IDX   = {k:i for i,k in enumerate(ITEMS)}

# Your priority items: bowel(9), eating(1), sitting/transfer(3)
KEY_BAD = ["Bowel", "Feed", "Transfer"]
KEY_BAD_IDX = [IDX[x] for x in KEY_BAD]

rng = np.random.default_rng(2026)

# -------------------------
# Column names (as in your dataset)
# -------------------------
COL = {
    "Age": "Age",
    "NIHSS": "NIHSS",
    "AspPNA": "AspirationPneumonia",
    "Sepsis": "Sepsis",
    "Delirium": "Delirium",
    "DepPHQ9": "ClinicalDepression_PHQ9",
    "NutriScore": "NutritionalStatus_score",
    "EarlyRehab": "EarlyRehabilitation",
    "EarlyMob": "EarlyMobilization",
    "Treatment": "Treatment",
}

def to_num(x, default=0.0):
    """Robust numeric conversion: handles NaN, bool, 'yes/no', '0/1', etc."""
    if pd.isna(x):
        return default
    if isinstance(x, (int, float, np.integer, np.floating)):
        return float(x)
    if isinstance(x, bool):
        return float(int(x))
    s = str(x).strip().lower()
    if s in {"yes", "y", "true", "t"}:
        return 1.0
    if s in {"no", "n", "false", "f"}:
        return 0.0
    try:
        return float(s)
    except Exception:
        return default

def safe_get(row, colname, default=0.0):
    return to_num(row[colname], default) if colname in row.index else default

def safe_probs(weights):
    """
    Convert arbitrary weights -> valid probability vector.
    If invalid, return uniform.
    """
    w = np.array(weights, dtype=float)
    w = np.nan_to_num(w, nan=0.0, posinf=0.0, neginf=0.0)
    w[w < 0] = 0.0
    s = w.sum()
    if not np.isfinite(s) or s <= 0:
        return np.ones_like(w) / len(w)
    return w / s

def enforce_hierarchy(v):
    """Stairs=2 only if Mobility>=2 and Transfer>=2"""
    v = v.copy()
    if not (v[IDX["Mobility"]] >= 2 and v[IDX["Transfer"]] >= 2):
        v[IDX["Stairs"]] = min(v[IDX["Stairs"]], 1)
    return v

def impairment_weights(row, adl_discharge_total):
    """
    Weights for allocating deficits (dropping from max).
    Strong bias: Bowel + Feed + Transfer.
    Uses your clinical features for realism, but never produces invalid weights.
    """
    w = np.ones(len(ITEMS), dtype=float)

    # Base disability pattern
    w[IDX["Mobility"]] += 1.6
    w[IDX["Transfer"]] += 1.2
    w[IDX["Toilet"]]   += 0.9
    w[IDX["Stairs"]]   += 0.8

    # YOUR priorities
    if adl_discharge_total <= 8:
        w[IDX["Bowel"]] += 5.0
        w[IDX["Feed"]] += 3.0
        w[IDX["Transfer"]] += 2.5
    elif adl_discharge_total <= 11:
        w[IDX["Bowel"]] += 3.0
        w[IDX["Feed"]] += 2.0
        w[IDX["Transfer"]] += 1.5
    else:
        # Mild group — DO NOT over-assign bowel impairment
        w[IDX["Bowel"]] += 0.8
        w[IDX["Feed"]] += 1.0
        w[IDX["Transfer"]] += 1.2

    if adl_discharge_total <= 8:
        w[IDX["Bowel"]]   += 2.0
        w[IDX["Feed"]]    += 1.0
        w[IDX["Bladder"]] += 1.2

    # Clinical features (robust numeric conversion)
    age   = safe_get(row, COL["Age"], 0)
    nihss = safe_get(row, COL["NIHSS"], 0)
    asp_pna = int(safe_get(row, COL["AspPNA"], 0))
    sepsis  = int(safe_get(row, COL["Sepsis"], 0))
    delir   = int(safe_get(row, COL["Delirium"], 0))
    phq9    = safe_get(row, COL["DepPHQ9"], 0)
    nutri   = safe_get(row, COL["NutriScore"], 999)
    early_rehab = int(safe_get(row, COL["EarlyRehab"], 0))
    early_mob   = int(safe_get(row, COL["EarlyMob"], 0))

    if age >= 75:
        w[IDX["Mobility"]] += 0.8
        w[IDX["Transfer"]] += 0.6
        w[IDX["Bowel"]]    += 0.4

    if nihss >= 10:
        w[IDX["Mobility"]] += 1.0
        w[IDX["Transfer"]] += 1.0
        w[IDX["Toilet"]]   += 0.7
        w[IDX["Feed"]]     += 0.6
        w[IDX["Bowel"]]    += 0.6
        w[IDX["Bladder"]]  += 0.4

    if asp_pna == 1:
        w[IDX["Feed"]]     += 0.8
    if sepsis == 1:
        w[IDX["Mobility"]] += 0.6
        w[IDX["Transfer"]] += 0.4
        w[IDX["Bowel"]]    += 0.4
        w[IDX["Bladder"]]  += 0.3
    if delir == 1:
        w[IDX["Toilet"]]   += 0.4
        w[IDX["Mobility"]] += 0.4
        w[IDX["Feed"]]     += 0.3

    if phq9 >= 10:
        w[IDX["Mobility"]] += 0.3
        w[IDX["Transfer"]] += 0.3
        w[IDX["Bowel"]]    += 0.2

    # If nutri score is low -> slightly worse
    if nutri <= 10:
        w[IDX["Mobility"]] += 0.3
        w[IDX["Transfer"]] += 0.2
        w[IDX["Bowel"]]    += 0.2

    # Early rehab/mobilization protects slightly (but never below 0)
    if early_rehab == 1:
        w[IDX["Mobility"]] -= 0.2
        w[IDX["Transfer"]] -= 0.2
    if early_mob == 1:
        w[IDX["Mobility"]] -= 0.2

    # Final safety: finite positive
    w = np.nan_to_num(w, nan=0.2, posinf=0.2, neginf=0.2)
    w = np.clip(w, 0.05, None)
    return w

def adjust_to_target_sum(v, target, deficit_weights):
    """Exact-sum adjustment with safe, deterministic order."""
    v = v.copy()
    target = int(target)

    add_order = [IDX["Groom"], IDX["Bath"], IDX["Dress"], IDX["Feed"], IDX["Toilet"],
                 IDX["Bladder"], IDX["Bowel"], IDX["Transfer"], IDX["Mobility"], IDX["Stairs"]]
    drop_order = list(np.argsort(-np.array(deficit_weights, float)))

    guard = 0
    while int(v.sum()) != target:
        guard += 1
        if guard > 20000:
            raise RuntimeError("Adjustment loop too long (check constraints).")

        if int(v.sum()) < target:
            for i in add_order:
                if v[i] < MAXS[i]:
                    v[i] += 1
                    break
            else:
                break
        else:
            for i in drop_order:
                if v[i] > MINS[i]:
                    v[i] -= 1
                    break
            else:
                break
    return v

def generate_discharge_items(row, target_total):
    """Start at max then drop deficits using safe probabilities."""
    target_total = int(target_total)
    v = MAXS.copy()
    need_drop = int(v.sum()) - target_total
    if need_drop < 0:
        raise ValueError(f"ADL_Discharge={target_total} > 20 is invalid")
    if need_drop == 0:
        return enforce_hierarchy(v)

    w = impairment_weights(row, target_total)

    guard = 0
    while need_drop > 0:
        guard += 1
        if guard > 20000:
            raise RuntimeError("Discharge generation loop too long.")

        droppable = np.where(v > MINS)[0]
        if len(droppable) == 0:
            break

        probs = safe_probs(w[droppable])
        i = rng.choice(droppable, p=probs)

        v[i] -= 1
        need_drop -= 1
        v = enforce_hierarchy(v)

    if int(v.sum()) != target_total:
        v = adjust_to_target_sum(v, target_total, w)
        v = enforce_hierarchy(v)
        if int(v.sum()) != target_total:
            v = adjust_to_target_sum(v, target_total, w)

    return v

def transition_to_6m(row, v_dis, target_6m_total, adl_dis_total):
    """
    Move from discharge to 6 months, exact-sum.
    KEY_BAD (Bowel/Feed/Transfer) are sticky if recovery small / low baseline / severe features.
    """
    v = v_dis.copy()
    target_6m_total = int(target_6m_total)
    delta = target_6m_total - int(v.sum())
    poor_recovery = abs(delta) <= 2
    low_baseline = int(adl_dis_total) <= 8

    nihss = safe_get(row, COL["NIHSS"], 0)
    sepsis = int(safe_get(row, COL["Sepsis"], 0))
    delir  = int(safe_get(row, COL["Delirium"], 0))
    asp_pna= int(safe_get(row, COL["AspPNA"], 0))
    phq9   = safe_get(row, COL["DepPHQ9"], 0)
    early_rehab = int(safe_get(row, COL["EarlyRehab"], 0))
    early_mob   = int(safe_get(row, COL["EarlyMob"], 0))

    p_keybad_improve = 0.55
    if poor_recovery: p_keybad_improve = 0.15
    if low_baseline:  p_keybad_improve *= 0.7
    if nihss >= 10:   p_keybad_improve *= 0.7
    if (sepsis or delir or asp_pna): p_keybad_improve *= 0.8
    if phq9 >= 10:    p_keybad_improve *= 0.85
    if (early_rehab or early_mob): p_keybad_improve = min(0.75, p_keybad_improve + 0.10)
    p_keybad_improve = float(np.clip(p_keybad_improve, 0.05, 0.90))

    improve_order = ["Mobility","Transfer","Toilet","Dress","Stairs","Feed","Bladder","Bowel","Groom","Bath"]
    decline_order = ["Mobility","Transfer","Toilet","Dress","Stairs","Feed","Bladder","Bowel","Groom","Bath"]
    improve_idx = [IDX[x] for x in improve_order]
    decline_idx = [IDX[x] for x in decline_order]

    def can_upgrade(i):
        if v[i] >= MAXS[i]:
            return False
        if i in KEY_BAD_IDX:
            return rng.random() < p_keybad_improve
        return True

    if delta > 0:
        for _ in range(delta):
            for i in improve_idx:
                if can_upgrade(i):
                    v[i] += 1
                    break
            else:
                candidates = np.where(v < MAXS)[0]
                if len(candidates) == 0: break
                weights = np.ones(len(ITEMS), float)
                if poor_recovery or low_baseline:
                    weights[KEY_BAD_IDX] *= 0.25
                cand = candidates
                probs = safe_probs(weights[cand])
                v[rng.choice(cand, p=probs)] += 1

    elif delta < 0:
        for _ in range(-delta):
            for i in decline_idx:
                if v[i] > MINS[i]:
                    v[i] -= 1
                    break
            else:
                candidates = np.where(v > MINS)[0]
                if len(candidates) == 0: break
                v[rng.choice(candidates)] -= 1

    w = impairment_weights(row, int(adl_dis_total))
    v = enforce_hierarchy(v)
    if int(v.sum()) != target_6m_total:
        v = adjust_to_target_sum(v, target_6m_total, w)
        v = enforce_hierarchy(v)
        if int(v.sum()) != target_6m_total:
            v = adjust_to_target_sum(v, target_6m_total, w)

    return v

# -------------------------
# Generate item-level columns
# -------------------------
dis_cols = [f"{x}_D" for x in ITEMS]
m6_cols  = [f"{x}_6M" for x in ITEMS]

dis_list, m6_list = [], []

for _, row in df.iterrows():
    t0 = int(to_num(row["ADL_Discharge"]))
    t1 = int(to_num(row["ADL_6month"]))

    v0 = generate_discharge_items(row, t0)
    v1 = transition_to_6m(row, v0, t1, t0)

    dis_list.append(v0)
    m6_list.append(v1)

dis_arr = np.vstack(dis_list)
m6_arr  = np.vstack(m6_list)

for j, c in enumerate(dis_cols):
    df[c] = dis_arr[:, j]
for j, c in enumerate(m6_cols):
    df[c] = m6_arr[:, j]

# -------------------------
# Sanity checks
# -------------------------
assert (df[dis_cols].sum(axis=1).astype(int) == df["ADL_Discharge"].astype(int)).all()
assert (df[m6_cols].sum(axis=1).astype(int)  == df["ADL_6month"].astype(int)).all()

# Optional diagnostics for your intended pattern:
df["delta_ADL"] = df["ADL_6month"] - df["ADL_Discharge"]
df["poor_recovery"] = (df["delta_ADL"] <= 2)
df["low_ADL"] = (df["ADL_Discharge"] <= 8)

subset = df[df["low_ADL"] & df["poor_recovery"]]
if len(subset) > 0:
    print("Low ADL + poor recovery: discharge impairment prevalence")
    print({
        "Bowel_D<=1": float((subset["Bowel_D"] <= 1).mean()),
        "Feed_D<=1": float((subset["Feed_D"] <= 1).mean()),
        "Transfer_D<=1": float((subset["Transfer_D"] <= 1).mean()),
    })

df.to_csv(OUT_PATH, index=False)
print("Saved:", OUT_PATH)
```

```python
import numpy as np
import pandas as pd

# ============================================================
# INPUT / OUTPUT
# ============================================================
IN_PATH  = "data_gen_with_ADL_items_logic_5.csv"
OUT_PATH = "data_gen_with_ADL_and_MNA_items_5.csv"

df = pd.read_csv(IN_PATH)
rng = np.random.default_rng(20260212)

# ============================================================
# Helper utilities (robust)
# ============================================================
def to_num(x, default=np.nan):
    if pd.isna(x):
        return default
    if isinstance(x, (int, float, np.integer, np.floating)):
        return float(x)
    if isinstance(x, bool):
        return float(int(x))
    s = str(x).strip().lower()
    if s in {"yes", "y", "true", "t"}:
        return 1.0
    if s in {"no", "n", "false", "f"}:
        return 0.0
    try:
        return float(s)
    except Exception:
        return default

def clip_int(x, lo, hi):
    return int(max(lo, min(hi, int(x))))

def safe_probs(w):
    w = np.array(w, dtype=float)
    w = np.nan_to_num(w, nan=0.0, posinf=0.0, neginf=0.0)
    w[w < 0] = 0.0
    s = w.sum()
    if (not np.isfinite(s)) or s <= 0:
        return np.ones_like(w) / len(w)
    return w / s

# ============================================================
# 0) Define recovery groups so "bad recovery" has higher prevalence of bad MNA items
#    (use delta_ADL if present, else compute)
# ============================================================
if "delta_ADL" not in df.columns:
    df["delta_ADL"] = df["ADL_6month"] - df["ADL_Discharge"]

# You can keep these cutoffs; they match what we used earlier
df["poor_recovery"] = (df["delta_ADL"] <= 2).astype(int)   # worse / slower recovery
df["fast_recovery"] = (df["delta_ADL"] >= 5).astype(int)   # good recovery

# ============================================================
# 1) MNA scoring rules (from your questionnaire image)
#    Screening (MNA-SF): A–F, max 14
#    Full MNA: Screening + Assessment (G–R), max 30
# ============================================================
# Screening items A–F max points per item
SCR_ITEMS = ["A", "B", "C", "D", "E", "F"]
SCR_MAX   = np.array([2,   3,   2,   2,   2,   3], dtype=int)  # D is 2 (no stress) or 0 (yes)
SCR_MIN   = np.zeros_like(SCR_MAX)
SCR_IDX   = {k:i for i,k in enumerate(SCR_ITEMS)}

# Assessment items G–R max points sum 16
# We'll store detailed subitems for K (3 yes/no) and compute K_score {0,0.5,1.0}
ASS_ITEMS = ["G","H","I","J","K","L","M","N","O","P","Q","R"]
ASS_MAX   = np.array([1,  1,  1,  2,  1,  1,  1,  2,  2,  2,  1,  1], dtype=float)
ASS_MIN   = np.zeros_like(ASS_MAX)
ASS_IDX   = {k:i for i,k in enumerate(ASS_ITEMS)}

SCR_TOTAL_MAX = int(SCR_MAX.sum())  # 14
ASS_TOTAL_MAX = float(ASS_MAX.sum())  # 16.0
FULL_TOTAL_MAX = 30.0

# ============================================================
# 2) Pull the "already listed" full MNA totals in your df
#    Per your note:
#      - nut_score = full MNA total at discharge (0–30)
#      - nut6      = full MNA total at 6 months (0–30)
#    If someone is normal and screening terminates (>=12), our generated full items may be NA
#    and total will NOT be forced to match nut_score/nut6 (by design).
# ============================================================
NUT_DIS_COL = "NutritionalStatus_score"
NUT_6M_COL  = "NutritionalStatus_6month_score"

if NUT_DIS_COL not in df.columns or NUT_6M_COL not in df.columns:
    raise ValueError("Expected columns 'nut_score' and 'nut6' were not found in the dataframe.")

# ============================================================
# 3) Clinical severity drivers for MNA generation
#    (no edits needed; uses columns if present, otherwise defaults)
# ============================================================
def row_val(row, name, default=0.0):
    return to_num(row[name], default) if name in row.index else default

def nutrition_severity(row, timepoint="D"):
    """
    Higher = worse nutrition risk.
    Ties to:
      - ADL severity
      - poor recovery
      - complications (if present)
    """
    if timepoint == "D":
        adl = row_val(row, "ADL_Discharge", 10)
    else:
        adl = row_val(row, "ADL_6month", 14)

    # base: lower ADL => higher risk
    sev = 0.0
    sev += (12 - adl) / 6.0  # roughly: ADL 6 -> +1.0, ADL 18 -> -1.0

    # recovery group effect
    poor = int(row_val(row, "poor_recovery", 0))
    fast = int(row_val(row, "fast_recovery", 0))
    sev += 0.6 * poor
    sev -= 0.2 * fast

    # complications (if present in df)
    sev += 0.4 * int(row_val(row, "AspirationPneumonia", 0))
    sev += 0.4 * int(row_val(row, "Sepsis", 0))
    sev += 0.3 * int(row_val(row, "Delirium", 0))

    # depression (if present)
    phq9 = row_val(row, "ClinicalDepression_PHQ9", 0)
    sev += 0.02 * max(0.0, phq9 - 9)  # gentle

    # early rehab/mobilization protective (if present)
    sev -= 0.15 * int(row_val(row, "EarlyRehabilitation", 0))
    sev -= 0.10 * int(row_val(row, "EarlyMobilization", 0))

    # noise
    sev += rng.normal(0, 0.15)
    return float(sev)

# ============================================================
# 4) Strategy: Generate Screening totals with feasibility for Full score matching
#    If Full is required and nut_total is available, we pick a screening total S such that:
#       nut_total - 16 <= S <= min(11, nut_total)
#    so the remaining assessment total fits 0..16.
# ============================================================
def choose_screen_total_for_full(nut_total):
    nut_total = float(nut_total)
    # Full total = Screening + Assessment, where Assessment in [0,16]
    # Screening must be in [0,11] if full is triggered (screen <=11)
    lo = max(0.0, nut_total - ASS_TOTAL_MAX)
    hi = min(11.0, nut_total)
    if lo > hi:
        # impossible; fallback to something feasible
        return int(min(11.0, max(0.0, nut_total)))
    # choose within [lo, hi] with mild bias toward middle
    s = rng.uniform(lo, hi)
    return int(round(s))

# ============================================================
# 5) Generate SCREENING item vector to match a chosen screening total
#    Clinically meaningful: low recovery => more severe intake decline, weight loss, low mobility, etc.
# ============================================================
def gen_screening_items(row, target_total=None, timepoint="D"):
    sev = nutrition_severity(row, timepoint)
    poor = int(row_val(row, "poor_recovery", 0))
    fast = int(row_val(row, "fast_recovery", 0))

    # Start from max then allocate deficits
    v = SCR_MAX.astype(int).copy()
    need_drop = int(v.sum()) - int(target_total) if target_total is not None else None

    # If target_total not provided, pick from severity:
    # - Better nutrition => higher score; worse => lower
    if target_total is None:
        # Map severity to expected screening total
        # sev ~ [-1, +2] typical
        mu = 12.5 - 2.5 * sev  # higher sev => lower score
        mu -= 0.6 * poor
        mu += 0.3 * fast
        s = int(round(np.clip(rng.normal(mu, 1.2), 0, 14)))
        target_total = s
        need_drop = int(v.sum()) - int(target_total)

    target_total = clip_int(target_total, 0, 14)
    need_drop = int(v.sum()) - target_total

    # Weights for which screening items get worse when dropping points:
    # YOUR goal: bad recovery -> more deficits in intake/weight/mobility/neuro
    # A: intake, B: weight loss, C: mobility, E: neuropsych, D: acute stress, F: BMI
    w = np.ones(len(SCR_ITEMS), float)
    w[SCR_IDX["A"]] += 2.0 + 1.0*poor
    w[SCR_IDX["B"]] += 1.8 + 0.8*poor
    w[SCR_IDX["C"]] += 1.5 + 0.5*poor
    w[SCR_IDX["E"]] += 1.0 + 0.4*poor
    w[SCR_IDX["D"]] += 0.7 + 0.2*poor
    w[SCR_IDX["F"]] += 0.8

    # Severe complications push intake/mobility down
    w[SCR_IDX["A"]] += 0.6 * int(row_val(row, "AspirationPneumonia", 0))
    w[SCR_IDX["C"]] += 0.4 * int(row_val(row, "Delirium", 0))
    w[SCR_IDX["E"]] += 0.3 * int(row_val(row, "ClinicalDepression_PHQ9", 0) >= 10)

    # Ensure positive
    w = np.clip(w, 0.05, None)

    # Drop points one by one respecting min
    guard = 0
    while need_drop > 0:
        guard += 1
        if guard > 20000:
            break
        droppable = np.where(v > SCR_MIN)[0]
        if len(droppable) == 0:
            break
        probs = safe_probs(w[droppable])
        i = rng.choice(droppable, p=probs)
        v[i] -= 1
        need_drop -= 1

    # Special rule for D (acute disease/stress): should be either 0 or 2
    # If it ended up as 1 due to point-dropping, snap it to nearest (0 or 2) and compensate elsewhere.
    if v[SCR_IDX["D"]] == 1:
        # choose 0 if poor recovery/complications else 2
        d_target = 0 if (poor == 1 or int(row_val(row, "Sepsis", 0)) == 1) else 2
        diff = d_target - v[SCR_IDX["D"]]
        v[SCR_IDX["D"]] = d_target
        # compensate diff by adjusting other items if needed
        while diff != 0:
            if diff > 0:
                # need add 1 somewhere else
                up = np.where(v < SCR_MAX)[0]
                up = up[up != SCR_IDX["D"]]
                if len(up) == 0:
                    break
                j = rng.choice(up)
                v[j] += 1
                diff -= 1
            else:
                down = np.where(v > SCR_MIN)[0]
                down = down[down != SCR_IDX["D"]]
                if len(down) == 0:
                    break
                j = rng.choice(down)
                v[j] -= 1
                diff += 1

    # Final exact sum fix (tiny)
    cur = int(v.sum())
    if cur != target_total:
        diff = target_total - cur
        # add or subtract with a simple order
        add_order = [SCR_IDX["F"], SCR_IDX["D"], SCR_IDX["E"], SCR_IDX["C"], SCR_IDX["B"], SCR_IDX["A"]]
        drop_order = [SCR_IDX["A"], SCR_IDX["B"], SCR_IDX["C"], SCR_IDX["E"], SCR_IDX["D"], SCR_IDX["F"]]
        guard = 0
        while diff != 0 and guard < 20000:
            guard += 1
            if diff > 0:
                moved = False
                for j in add_order:
                    if v[j] < SCR_MAX[j] and not (j == SCR_IDX["D"] and v[j] == 0):  # allow D 0->2 only via jump below
                        # D must jump 0<->2
                        if j == SCR_IDX["D"]:
                            if v[j] == 0 and diff >= 2:
                                v[j] = 2
                                diff -= 2
                                moved = True
                                break
                            continue
                        v[j] += 1
                        diff -= 1
                        moved = True
                        break
                if not moved:
                    break
            else:
                moved = False
                for j in drop_order:
                    if v[j] > SCR_MIN[j]:
                        if j == SCR_IDX["D"]:
                            if v[j] == 2 and diff <= -2:
                                v[j] = 0
                                diff += 2
                                moved = True
                                break
                            continue
                        v[j] -= 1
                        diff += 1
                        moved = True
                        break
                if not moved:
                    break

    # If D is not 0/2, snap again
    if v[SCR_IDX["D"]] not in (0, 2):
        v[SCR_IDX["D"]] = 0 if poor == 1 else 2

    return v

def screening_score(v):
    return int(np.sum(v))

# ============================================================
# 6) Generate FULL ASSESSMENT items G–R to match target assessment total (0..16)
#    Clinical meaning: low recovery => more pressure sores, feeding difficulty, low fluids, low CC/MAC, etc.
# ============================================================
def k_from_yescount(c):
    # K score: 0 if 0 or 1 yes; 0.5 if 2 yes; 1.0 if 3 yes
    if c >= 3:
        return 1.0
    if c == 2:
        return 0.5
    return 0.0

def gen_k_subitems_from_score(k_score, poor_bias=False):
    """
    Generate (K_dairy, K_legumes_eggs, K_meat_fish_poultry) as 0/1 yes/no.
    If poor_bias=True, tends to have fewer yes.
    """
    if k_score >= 1.0:
        return (1, 1, 1)
    if abs(k_score - 0.5) < 1e-9:
        # exactly two yes
        choices = [(1,1,0),(1,0,1),(0,1,1)]
        if poor_bias:
            # still 2 yes, but random which one is missing
            return choices[rng.integers(0, len(choices))]
        return choices[rng.integers(0, len(choices))]
    # k_score = 0.0 -> 0 or 1 yes
    if poor_bias and rng.random() < 0.7:
        return (0,0,0)
    # otherwise one yes
    choices = [(1,0,0),(0,1,0),(0,0,1),(0,0,0)]
    return choices[rng.integers(0, len(choices))]

def gen_assessment_items(row, scr_vec, target_ass_total, timepoint="D"):
    """
    scr_vec already fixed.
    target_ass_total must be feasible 0..16.
    Returns:
      - ass_vec (length 12: G..R with K as numeric score)
      - k_subitems dict (3 yes/no)
    """
    poor = int(row_val(row, "poor_recovery", 0))
    fast = int(row_val(row, "fast_recovery", 0))
    sev  = nutrition_severity(row, timepoint)

    target_ass_total = float(np.clip(target_ass_total, 0.0, ASS_TOTAL_MAX))

    # Start from max then drop deficits
    v = ASS_MAX.copy()
    need_drop = float(v.sum() - target_ass_total)

    # Weights: which assessment items are likely to be impaired
    # Low recovery => higher deficits in pressure sores, feeding, fluids, CC/MAC, protein/veg
    w = np.ones(len(ASS_ITEMS), float)

    # Frailty / sarcopenia cluster
    w[ASS_IDX["R"]] += 1.4 + 0.8*poor   # calf circumference
    w[ASS_IDX["Q"]] += 1.2 + 0.6*poor   # MAC
    w[ASS_IDX["I"]] += 1.0 + 0.7*poor   # pressure sores
    w[ASS_IDX["G"]] += 0.8 + 0.4*poor   # independent living

    # Intake quality cluster
    w[ASS_IDX["J"]] += 1.2 + 0.6*poor   # meals/day
    w[ASS_IDX["K"]] += 1.0 + 0.6*poor   # protein markers
    w[ASS_IDX["L"]] += 0.9 + 0.5*poor   # fruit/veg
    w[ASS_IDX["M"]] += 0.8 + 0.5*poor   # fluids
    w[ASS_IDX["N"]] += 0.9 + 0.6*poor   # mode of feeding

    # Health / perception
    w[ASS_IDX["O"]] += 0.4 + 0.2*poor   # self view nutrition
    w[ASS_IDX["P"]] += 0.4 + 0.2*poor   # health comparison
    w[ASS_IDX["H"]] += 0.3              # polypharmacy often present but not always severe

    # Tie to screening mobility and intake for coherence:
    # If screening mobility is poor (C low), make independence and CC/MAC more likely low.
    C = scr_vec[SCR_IDX["C"]]
    A = scr_vec[SCR_IDX["A"]]
    if C <= 1:
        w[ASS_IDX["G"]] += 0.8
        w[ASS_IDX["R"]] += 0.6
        w[ASS_IDX["Q"]] += 0.4
    if A <= 1:
        w[ASS_IDX["J"]] += 0.6
        w[ASS_IDX["K"]] += 0.6
        w[ASS_IDX["N"]] += 0.5
        w[ASS_IDX["M"]] += 0.3

    # Complications
    w[ASS_IDX["N"]] += 0.5 * int(row_val(row, "AspirationPneumonia", 0))
    w[ASS_IDX["I"]] += 0.5 * int(row_val(row, "Sepsis", 0))
    w[ASS_IDX["O"]] += 0.2 * int(row_val(row, "Delirium", 0))

    # Fast recovery slightly protects
    w -= 0.15 * fast
    w = np.clip(w, 0.05, None)

    # Drop in steps of 0.5 or 1 depending on item:
    # Allowed steps:
    # G/H/I/L/R: 1
    # J: 1
    # K: 0.5
    # M/P/Q: 0.5
    # N/O: 1
    step = {
        "G":1.0,"H":1.0,"I":1.0,"J":1.0,"K":0.5,"L":1.0,"M":0.5,
        "N":1.0,"O":1.0,"P":0.5,"Q":0.5,"R":1.0
    }

    guard = 0
    while need_drop > 1e-9 and guard < 30000:
        guard += 1
        droppable = []
        droppable_w = []
        for item in ASS_ITEMS:
            i = ASS_IDX[item]
            s = step[item]
            if v[i] - s >= ASS_MIN[i] - 1e-9:
                droppable.append(i)
                droppable_w.append(w[i])
        if len(droppable) == 0:
            break

        probs = safe_probs(droppable_w)
        i = rng.choice(droppable, p=probs)

        # drop one step if it doesn't overshoot too badly
        item = ASS_ITEMS[i]
        s = step[item]
        # If remaining drop is smaller than step, we may still drop and fix later
        v[i] = round(float(v[i] - s), 3)
        need_drop = float(v.sum() - target_ass_total)

    # Coherence constraints (post):
    # - If screening mobility very poor, don't allow G=1 (independent) too often
    if scr_vec[SCR_IDX["C"]] == 0 and v[ASS_IDX["G"]] == 1.0 and rng.random() < 0.75:
        v[ASS_IDX["G"]] = 0.0

    # - If feeding difficulty likely (A low), N should not be 2 too often
    if scr_vec[SCR_IDX["A"]] == 0 and v[ASS_IDX["N"]] == 2.0 and rng.random() < 0.8:
        v[ASS_IDX["N"]] = 1.0

    # Fix exact sum if needed (adjust using small-step items first: K, M, P, Q)
    def adjust_exact(v, target):
        v = v.copy()
        target = float(target)
        small_items = ["K","M","P","Q"]
        big_items   = ["J","N","O","G","H","I","L","R"]

        guard2 = 0
        while abs(float(v.sum() - target)) > 1e-6 and guard2 < 30000:
            guard2 += 1
            diff = float(target - v.sum())
            if diff > 0:
                # add points
                moved = False
                # try small items first (0.5 steps)
                for it in small_items + big_items:
                    i = ASS_IDX[it]
                    s = step[it]
                    if v[i] + s <= ASS_MAX[i] + 1e-9 and (diff >= s - 1e-9):
                        v[i] = round(float(v[i] + s), 3)
                        moved = True
                        break
                if not moved:
                    break
            else:
                # subtract points
                moved = False
                for it in small_items + big_items:
                    i = ASS_IDX[it]
                    s = step[it]
                    if v[i] - s >= ASS_MIN[i] - 1e-9 and (abs(diff) >= s - 1e-9):
                        v[i] = round(float(v[i] - s), 3)
                        moved = True
                        break
                if not moved:
                    break
        return v

    v = adjust_exact(v, target_ass_total)

    # Build K subitems consistent with K score
    K_score = float(v[ASS_IDX["K"]])
    # snap K to valid {0,0.5,1}
    if K_score not in (0.0, 0.5, 1.0):
        K_score = min((0.0, 0.5, 1.0), key=lambda z: abs(z - K_score))
        v[ASS_IDX["K"]] = K_score

    k_dairy, k_leg_egg, k_meat = gen_k_subitems_from_score(
        K_score,
        poor_bias=(poor == 1 or sev > 0.8)
    )

    # Verify K_score matches yescount rule (force if needed)
    yescount = k_dairy + k_leg_egg + k_meat
    forced_k = k_from_yescount(yescount)
    if forced_k != K_score:
        # force subitems to match K_score
        k_dairy, k_leg_egg, k_meat = gen_k_subitems_from_score(K_score, poor_bias=(poor == 1))
        yescount = k_dairy + k_leg_egg + k_meat
        forced_k = k_from_yescount(yescount)

    k_sub = {"K_dairy": k_dairy, "K_legumes_eggs": k_leg_egg, "K_meat_fish_poultry": k_meat}

    # Final exact correction again (in case post-hoc coherence changed sum)
    v = adjust_exact(v, target_ass_total)

    return v, k_sub

# ============================================================
# 7) Main generation per timepoint
#    - If screening >= 12: terminate -> Full items set NA; full total not forced to match nut_score/nut6
#    - If screening <= 11: generate Full items and FORCE full total to match nut_score/nut6 (when feasible)
# ============================================================
def generate_mna_timepoint(row, nut_total, timepoint="D"):
    """
    Returns dict of MNA item answers for one timepoint:
      - Screening A..F, ScreeningTotal
      - FullRequired flag (screen <=11)
      - If FullRequired: G..R + K subitems + FullTotal (forced to nut_total if feasible)
      - If not required: G..R are NaN
    """
    out = {}

    nut_total = to_num(nut_total, default=np.nan)

    # Decide if we will try to match full total:
    match_full = np.isfinite(nut_total) and (0 <= nut_total <= 30)

    # If we need full matching, choose a feasible screening total that allows assessment in 0..16
    if match_full:
        scr_target = choose_screen_total_for_full(nut_total)
        # scr_target must be <=11 to trigger full; if it comes out 12..14 we treat it as "screen terminates"
        # and will not generate full items (consistent with your note).
        scr_vec = gen_screening_items(row, target_total=scr_target, timepoint=timepoint)
    else:
        scr_vec = gen_screening_items(row, target_total=None, timepoint=timepoint)

    # Save A..F
    for i, k in enumerate(SCR_ITEMS):
        out[f"MNA_{k}_{timepoint}"] = int(scr_vec[i])

    scr_total = screening_score(scr_vec)
    out[f"MNA_SF_Total_{timepoint}"] = int(scr_total)

    # Rule: if SF score >= 12, terminate (no full)
    full_required = int(scr_total <= 11)
    out[f"MNA_FullRequired_{timepoint}"] = full_required

    if full_required == 0:
        # terminate: do NOT attempt to match nut_total (per your instruction)
        for k in ASS_ITEMS:
            out[f"MNA_{k}_{timepoint}"] = np.nan
        out[f"MNA_Assess_Total_{timepoint}"] = np.nan
        out[f"MNA_Full_Total_Gen_{timepoint}"] = float(scr_total)  # only screening exists
        # K subitems NA
        out[f"MNA_K_dairy_{timepoint}"] = np.nan
        out[f"MNA_K_legumes_eggs_{timepoint}"] = np.nan
        out[f"MNA_K_meat_fish_poultry_{timepoint}"] = np.nan
        return out

    # Full required -> if nut_total exists, force full total to it; otherwise generate a plausible total
    if match_full:
        target_full = float(np.clip(nut_total, 0.0, 30.0))
    else:
        # If no given total, generate plausible full total: screening + assessment depending on severity
        sev = nutrition_severity(row, timepoint)
        mu = 24.0 - 4.0*sev - 1.5*int(row_val(row, "poor_recovery", 0))
        target_full = float(np.clip(rng.normal(mu, 3.0), 8.0, 28.0))

        # Ensure full_required implies SF <= 11
        if scr_total > 11:
            # force it down a bit
            scr_total = 11
            # regenerate screening to exactly 11
            scr_vec = gen_screening_items(row, target_total=11, timepoint=timepoint)
            for i, k in enumerate(SCR_ITEMS):
                out[f"MNA_{k}_{timepoint}"] = int(scr_vec[i])
            out[f"MNA_SF_Total_{timepoint}"] = 11
            full_required = 1
            out[f"MNA_FullRequired_{timepoint}"] = 1

    # Compute target assessment total
    # Need in [0,16]; if not feasible because SF is too low/high, we adjust by recomputing SF target within feasibility
    target_ass = target_full - float(out[f"MNA_SF_Total_{timepoint}"])
    if target_ass < 0.0 or target_ass > ASS_TOTAL_MAX + 1e-6:
        # re-pick screening total to make it feasible
        scr_target = choose_screen_total_for_full(target_full)
        scr_target = min(scr_target, 11)
        scr_vec = gen_screening_items(row, target_total=scr_target, timepoint=timepoint)
        for i, k in enumerate(SCR_ITEMS):
            out[f"MNA_{k}_{timepoint}"] = int(scr_vec[i])
        scr_total = screening_score(scr_vec)
        out[f"MNA_SF_Total_{timepoint}"] = int(scr_total)
        out[f"MNA_FullRequired_{timepoint}"] = int(scr_total <= 11)

        target_ass = target_full - float(scr_total)
        target_ass = float(np.clip(target_ass, 0.0, ASS_TOTAL_MAX))

    # Generate assessment items to match target_ass
    ass_vec, k_sub = gen_assessment_items(row, scr_vec, target_ass_total=target_ass, timepoint=timepoint)

    # Save G..R (K is numeric score)
    for i, k in enumerate(ASS_ITEMS):
        out[f"MNA_{k}_{timepoint}"] = float(ass_vec[i])

    # Save K subitems
    out[f"MNA_K_dairy_{timepoint}"] = int(k_sub["K_dairy"])
    out[f"MNA_K_legumes_eggs_{timepoint}"] = int(k_sub["K_legumes_eggs"])
    out[f"MNA_K_meat_fish_poultry_{timepoint}"] = int(k_sub["K_meat_fish_poultry"])

    out[f"MNA_Assess_Total_{timepoint}"] = float(np.sum(ass_vec))
    out[f"MNA_Full_Total_Gen_{timepoint}"] = float(out[f"MNA_SF_Total_{timepoint}"] + out[f"MNA_Assess_Total_{timepoint}"])

    return out

# ============================================================
# 8) Run generation for discharge and 6 months
# ============================================================
rows_out = []
for _, row in df.iterrows():
    out_d  = generate_mna_timepoint(row, row[NUT_DIS_COL], timepoint="D")
    out_6m = generate_mna_timepoint(row, row[NUT_6M_COL],  timepoint="6M")
    merged = {**out_d, **out_6m}
    rows_out.append(merged)

mna_items = pd.DataFrame(rows_out)
df2 = pd.concat([df, mna_items], axis=1)

for tp in ["D", "6M"]:
    poor = (df2["poor_recovery"] == 1)
    fast = (df2["fast_recovery"] == 1)

    # Screening items (always exist)
    df2.loc[poor, f"MNA_A_{tp}"] = np.clip(
        df2.loc[poor, f"MNA_A_{tp}"] - (rng.random(poor.sum()) < 0.10).astype(int),
        0, 2
    )
    df2.loc[fast, f"MNA_A_{tp}"] = np.clip(
        df2.loc[fast, f"MNA_A_{tp}"] + (rng.random(fast.sum()) < 0.10).astype(int),
        0, 2
    )

    df2.loc[poor, f"MNA_C_{tp}"] = np.clip(
        df2.loc[poor, f"MNA_C_{tp}"] - (rng.random(poor.sum()) < 0.10).astype(int),
        0, 2
    )
    df2.loc[fast, f"MNA_C_{tp}"] = np.clip(
        df2.loc[fast, f"MNA_C_{tp}"] + (rng.random(fast.sum()) < 0.10).astype(int),
        0, 2
    )

    # Full items only if FullRequired==1
    full_poor = poor & (df2[f"MNA_FullRequired_{tp}"] == 1)
    full_fast = fast & (df2[f"MNA_FullRequired_{tp}"] == 1)

    # I (pressure sore): 0=yes (worse), 1=no (better)
    df2.loc[full_poor, f"MNA_I_{tp}"] = np.where(
        rng.random(full_poor.sum()) < 0.10, 0.0, df2.loc[full_poor, f"MNA_I_{tp}"]
    )
    df2.loc[full_fast, f"MNA_I_{tp}"] = np.where(
        rng.random(full_fast.sum()) < 0.10, 1.0, df2.loc[full_fast, f"MNA_I_{tp}"]
    )

    # N (feeding mode): worse = <=1
    df2.loc[full_poor, f"MNA_N_{tp}"] = np.clip(
        df2.loc[full_poor, f"MNA_N_{tp}"] - (rng.random(full_poor.sum()) < 0.15).astype(int),
        0, 2
    )
    df2.loc[full_fast, f"MNA_N_{tp}"] = np.clip(
        df2.loc[full_fast, f"MNA_N_{tp}"] + (rng.random(full_fast.sum()) < 0.15).astype(int),
        0, 2
    )

    # R (calf circumference): 0=low (worse), 1=normal (better)
    df2.loc[full_poor, f"MNA_R_{tp}"] = np.where(
        rng.random(full_poor.sum()) < 0.25, 0.0, df2.loc[full_poor, f"MNA_R_{tp}"]
    )
    df2.loc[full_fast, f"MNA_R_{tp}"] = np.where(
        rng.random(full_fast.sum()) < 0.25, 1.0, df2.loc[full_fast, f"MNA_R_{tp}"]
    )

# Recompute Screening totals after modifying A/C
df2["MNA_SF_Total_D"]  = (df2["MNA_A_D"]  + df2["MNA_B_D"]  + df2["MNA_C_D"]  + df2["MNA_D_D"]  + df2["MNA_E_D"]  + df2["MNA_F_D"]).astype(int)
df2["MNA_SF_Total_6M"] = (df2["MNA_A_6M"] + df2["MNA_B_6M"] + df2["MNA_C_6M"] + df2["MNA_D_6M"] + df2["MNA_E_6M"] + df2["MNA_F_6M"]).astype(int)

# ---------- NOW PRINT PREVALENCE (AFTER PATCH) ----------
def prevalence_worse(group_df, tp):
    res = {}
    if len(group_df) == 0:
        return res
    res[f"A<=1_{tp}"] = float((group_df[f"MNA_A_{tp}"] <= 1).mean())
    res[f"B<=1_{tp}"] = float((group_df[f"MNA_B_{tp}"] <= 1).mean())
    res[f"C<=1_{tp}"] = float((group_df[f"MNA_C_{tp}"] <= 1).mean())
    full = group_df[group_df[f"MNA_FullRequired_{tp}"] == 1]
    if len(full) > 0:
        res[f"I_pressure_yes_{tp}"] = float((full[f"MNA_I_{tp}"] == 0).mean())
        res[f"N<=1_{tp}"] = float((full[f"MNA_N_{tp}"] <= 1).mean())
        res[f"R_low_calf_{tp}"] = float((full[f"MNA_R_{tp}"] == 0).mean())
    return res

poor = df2[df2["poor_recovery"] == 1]
fast = df2[df2["fast_recovery"] == 1]

print("=== MNA worse-feature prevalence (Poor recovery) [AFTER PATCH] ===")
print("Discharge:", prevalence_worse(poor, "D"))
print("6 months :", prevalence_worse(poor, "6M"))

print("\n=== MNA worse-feature prevalence (Fast recovery) [AFTER PATCH] ===")
print("Discharge:", prevalence_worse(fast, "D"))
print("6 months :", prevalence_worse(fast, "6M"))

# ---------- SAVE LAST ----------
df2.to_csv(OUT_PATH, index=False)
print("\nSaved:", OUT_PATH)
```
